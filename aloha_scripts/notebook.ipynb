```python
import cv2
import zarr
import numpy as np
import matplotlib.pyplot as plt

def plot(zarr_file_path):
    # Open the Zarr file in read mode ('r')
    zarr_file = zarr.open(zarr_file_path, mode='r')

    positions = zarr_file['episode_0']['position']
    

    num_joints = positions.shape[1]
    time_steps = range(len(positions))

    # Plot each joint position over time
    plt.figure(figsize=(12, 8))
    # for joint_idx in range(num_joints):
    #     plt.plot(time_steps, positions[:, joint_idx], label=f'Joint {joint_idx + 1}')

    plt.plot(time_steps, positions[:, 4], label=f'Joint {5}')
    plt.title(f'Joint Positions for episiode {0}')
    plt.xlabel('Time Step')
    plt.ylabel('Position')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Show the plot for each episode
    plt.show()

if __name__ == '__main__':
    # Define the path to the Zarr file
    zarr_file_path = '/home/kyutae/test.zarr'

    # Call the function to play the video
    plot(zarr_file_path)
```
